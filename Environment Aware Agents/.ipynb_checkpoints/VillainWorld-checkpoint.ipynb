```python
import random
from copy import deepcopy
#from goto import with_goto

class Agent:
    def __init__(self):
        self.goal = ['',''] #G,D
        self.curr = [0,0] #start for now
        self.x = [1,-1,0,0] #up, down, L, R
        self.y = [0,0,-1,1]
        self.path = []
        self.col = 3
        self.row = 5
        self.plist ={} # dict
        self.vlist ={}
        self.glist ={}
        # initialise the p,v,g lists
        for i in range(4): #up, down, L, R
            self.plist[i] = []
            self.vlist[i] = []
            self.glist[i] = []

        # initiatlise the location of P, V, G, D
        self.p = []
        self.v = []
        self.g = []
        self.d = []

        self.visited = []
        self.env = []
        # initialise env and visited array 6*4
        for i in range(self.row+1):
            l = []; r = []
            for j in range(self.col+1):
                l.append(0)
                r.append('e') # empty cell
            self.visited.append(l)
            self.env.append(r)

        self.visited[0][0] = 1 #start pos
        self.setEnv()

    # given in question
    def setEnv(self):
        # set gold env
        self.env[5][0] = self.env[5][2] = self.env[4][1] = 'g'
        self.env[5][1] = 'G'

        # set pit env
        self.env[4][2] = self.env[5][3] = self.env[3][3] = 'b'
        self.env[4][3] = 'P'

        # set villain env 
        self.env[3][0] = self.env[2][1] = self.env[3][2] = 's'
        self.env[4][1] += 's'
        self.env[3][1] = 'V'

        # set destination
        self.env[0][3] = 'D'

        for i in range(self.row+1):
            print(self.env[i])


    def checkgoal(self):
        if (self.goal[0]=="G" and self.goal[1]=="D"):
            return 1
        else: 
            return 0

    def generate_move(self):
        possible = []
        for i in range(len(self.x)):
            x = self.curr[0]+self.x[i]
            y = self.curr[1]+self.y[i]
            if (x in range(0,self.row+1) and y in range(0,self.col+1) and self.visited[x][y]<20):
                possible.append([x,y])
        return possible
    
    def generate_move_obs(self,curr):
        possible = []
        for i in range(len(self.x)):
            x = curr[0]+self.x[i]
            y = curr[1]+self.y[i]
            if (x in range(0,self.row+1) and y in range(0,self.col+1) and self.visited[x][y]==0):
                possible.append([x,y])
        print("risk list",possible)
        return possible

    #@with_goto
    def check_obstacle(self,curr):  ########################
        x = curr[0]; y = curr[1]
        # check for e
        if (self.env[x][y]=='e'):
            print("Empty cell")
            return 0
        # check for P, V
        if (self.env[x][y]=='P'):
            return 1
        if (self.env[x][y]=='V'):
            return 1
        # check for G, D
        if (self.env[x][y]=='G'):
            print("Gold found")
            self.goal[0] = 'G' # taking gold
            return 0
        if (self.env[x][y]=='D'):
            print("Destination")
            if (self.goal[0]=='G'):
                self.goal[1] = 'D'
            return 0


        # pit
        if (self.env[x][y] == 'b' and len(self.p)==0):
            print("Breeze")
            safelist = self.generate_move_obs(curr) # for possible P pos
            # find which index of plist has empty list
            k = 0 # store index 
            f = 1
            for j in range(len(self.plist)):
                if safelist!=self.plist[j] and (not self.plist[j]):
                    k = j
                    f=0
                    break
                
            if f==0:
                for i in safelist:
                    self.plist[k].append(i)

                pans =[]
                for j in range(len(self.plist)):
                    for z in range(len(self.plist)):
                        if z==j:
                            continue
                        else:
                            for i in range(len(self.plist)):
                                try:
                                    if (self.plist[j][i]) in (self.plist[z]):
                                        pans = self.plist[j][i]
                                        break
                                    elif (self.plist[z][i]) in (self.plist[j]):
                                        pans = self.plist[z][i]
                                        break
                                except IndexError:
                                    continue
                if len(pans)!=0: # P found
                    self.p = pans
                    print("P = ",self.p)
                    print("P found")
                    return 1
            # if its seen for the first time
            return 1

        elif (self.env[x][y] == 'b' and len(self.p)): # ignoring b's when P found
            return 0

        # villain 
        if (self.env[x][y] == 's' and len(self.v)==0):
            print("Stench")
            safelist = self.generate_move_obs(curr) # for possible V pos
            # find which index of vlist has empty list
            k = 0 # store index 
            f=1
            for j in range(len(self.vlist)):
                if (not self.vlist[j]) and safelist!=self.vlist[j]:
                    k = j
                    f=0
                    break
                
            if f==0:
                for i in safelist:
                    self.vlist[k].append(i)

                vans = []
                for j in range(len(self.vlist)):
                    for z in range(len(self.vlist)):
                        if z==j:
                            continue
                        else:
                            for i in range(len(self.vlist)):
                                try:
                                    if (self.vlist[j][i]) in (self.vlist[z]):
                                        vans = self.vlist[j][i]
                                        break
                                    elif (self.vlist[z][i]) in (self.vlist[j]):
                                        vans = self.vlist[z][i]
                                        break
                                except IndexError:
                                    continue
                        
                if len(vans)!=0: # V found
                    self.v = vans 
                    print("V = ",self.v)
                    print("V found")
                    return 1
            # if its seen for the first time
            return 1

        elif (self.env[x][y] == 's' and len(self.v)): # ignoring s' when V found
            return 0

        # gold
        if (self.env[x][y] == 'g' and len(self.g)==0):
            print("Glitter")
            safelist = self.generate_move_obs(curr) # for possible G pos
            # find which index of glist has empty list
            k = 0 # store index 
            f=1
            for j in range(len(self.glist)):
                if (not self.glist[j]) and safelist!=self.glist[j]:
                    k = j
                    f=0
                    break
                
            if f==0:
                for i in safelist:
                    self.glist[k].append(i)

                gans = []
                for j in range(len(self.glist)):
                    for z in range(len(self.glist)):
                        if z==j:
                            continue
                        else:
                            for i in range(len(self.glist)):
                                try: 
                                    if (self.glist[j][i]) in (self.glist[z]):
                                        gans =self.glist[j][i]
                                        break
                                    elif (self.glist[z][i]) in (self.glist[j]):
                                        gans=self.glist[z][i]
                                        break
                                except IndexError:
                                    continue

                if len(gans)!=0: # G found
                    self.g = gans 
                    print("G = ",self.g)
                    print("G found")
                    return 1
                # if its seen for the first time
                return 1

        elif (self.env[x][y] == 'g' and len(self.g)): # ignoring g's when G found
            return 0

        # gs
        if (self.env[x][y]=='gs'):
            print("Stench and glitter")
            if (len(self.v) and len(self.g)):
                return 0 # ignore
            elif (len(self.v) and len(self.g)==0):
                safelist = self.generate_move_obs(curr) # for possible G pos
                # find which index of glist has empty list
                f=1
                k = 0 # store index 
                for j in range(len(self.glist)):
                    if (not self.glist[j]) and safelist!=self.glist[j]:
                        k = j
                        f=0
                        break
                if f==0:
                    for i in safelist:
                        self.glist[k].append(i)

                    gans = []
                    for j in range(len(self.glist)):
                        for z in range(len(self.glist)):
                            if z==j:
                                continue
                            else:
                                for i in range(len(self.glist)):
                                    try: 
                                        if (self.glist[j][i]) in (self.glist[z]):
                                            gans =self.glist[j][i]
                                            break
                                        elif (self.glist[z][i]) in (self.glist[j]):
                                            gans=self.glist[z][i]
                                            break
                                    except IndexError:
                                        continue
                            
                    if len(gans)!=0: # G found
                        self.g = gans 
                        print("G = ",self.p)
                        print("G found")
                        return 1
                    # if its seen for the first time
                    return 1



    def move(self):
        num=0
        moves = [self.curr]
        while (True and moves):
            moves = self.generate_move()
            random.shuffle(moves)
            print("Safe moves: ",moves)
            for i in range(len(moves)):
                curr = deepcopy(moves[i]) #new move
                if(self.check_obstacle(curr)):
                        # backtrack to previous
                        # make a different move
                    print("Move is ", self.curr, "and obstacle there\n")
                    for i in range(len(self.path)):
                        try:
                            temp = self.curr
                            back = self.path[len(self.path)-i-1]
                            if back == temp:
                                continue
                            else:
                                self.curr = deepcopy(back)
                                self.path.append(self.curr)
                                self.visited[self.curr[0]][self.curr[1]] += 1
                                print("Move is ",self.curr,"\n")
                             
                        except IndexError:
                            continue


                else:
                    self.curr = deepcopy(curr)
                    self.visited[self.curr[0]][self.curr[1]] += 1
                    print("Move is ",self.curr,"\n")
                    self.path.append(self.curr)
                    break # breaks out of for loop
            if(self.checkgoal()):
                    print("Success!!!")
                    #print(self.path)
                    break 
            num+=1
        else:
            print("Took ",num, " steps")
            print("Can't move from here and target not reached")
        
```

```python
goldAgent = Agent()
```

```python
goldAgent.move()
```
