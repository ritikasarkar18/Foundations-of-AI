```python
import random
```

```python
class Env_Aware_Agent:
    def __init__(self):
        self.pos = 0
        
    def move(self,env):
        num = env.get_limit()
        print("LIMIT is ",num)
        while self.pos<num:
            prop = env.get_property(self.pos)
            self.curr_pos()
            if prop == "O":
                print("The next position is O")
                print("Have to jump from ", self.pos)
                self.pos+=2
                print("The current position after jumping is ", self.pos)
            elif prop =="F":
                print("The next position is F")
                print("Going to walk from ", self.pos)
                self.pos+=1
                print("The current position after walking is ", self.pos)
                
            if self.pos==num-1:
                print("Reached the end of the road")
            
            
    def curr_pos(self):
        print("\nThe current position is ", self.pos)
```

```python
class Environment:
    def __init__(self):
        self.limit = 60
        self.property = {}
        self.env = ["O","F"]
        pos = 0
        while pos<self.limit:
            self.property[pos] = random.choice(self.env)
            
            if self.property[pos] == self.env[0]:
                self.property[pos+1] = self.env[1]
                pos+=2
            
            else:
                pos+=1
            
    def get_limit(self):
        return self.limit
    
    def get_property(self,pos):
        return self.property[pos]
        
```

```python
agent = Env_Aware_Agent()
env = Environment()

agent.move(env)
```
