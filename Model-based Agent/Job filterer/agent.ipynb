```python
import pandas as pd
```

```python
class JobPortal:
    def __init__(self):
        self.limit = [2,2,1,1,1] # no of candidates to be admitted
        
        self.jobstatus = {'SA':0,'SSA':0,'PL':0,'AVP':0,'VP':0}
        self.jobroles = list(self.jobstatus.keys())


    #more functions
    def get_jobroles(self):
        return self.jobroles

    def get_status(self,jname):
        return self.jobstatus[jname]

    def get_limit(self):
        return self.limit

    def update_vacancy(self, i):
        self.limit[i]-=1
        
    def update_status(self, i):
        self.jobstatus[i] += 1
```

```python
class ModelAgent:
    def __init__(self,env):
        self.stop = env.get_limit()
        self.prereq = {
            'SA':[['Software Developer','Software Engineer','Manager'],2],
            'SSA':[['Software Analyst','Senior Manager'],3],
            'PL':[['Senior Software Analyst','Senior Software Engineer'],5],
            'AVP':[['Chief Operating Officer','Senior Project Manager'],8],
            'VP':[['Assistant Vice President'],10]
        }
        self.black = ['Deloitte','Cognizant'] #blacklisted companies
        self.printStatus(env,list(self.prereq.keys()))

    def filter(self,env,name,userinput):
        roles = env.get_jobroles()
        
#         print(userinput)
        numofjobs = len(userinput)
        jobtypes = [i[1] for i in userinput]
        years = [i[0] for i in userinput]
        company = [i[2] for i in userinput]
        message = "Sorry you are not qualified\n"
        
        #checks
        matched=[]
        for i in range(len(jobtypes)):
            #check roles
            for j in roles:
                if jobtypes[i] in self.prereq[j][0]:
                    matched.append([j,years[i],company[i]]) #matched job
            # check years
#             print(matched)
        if (matched):
            y = [0]*len(self.prereq)
            for i in range(len(matched)):
                for j in range(len(roles)):
#                 print(self.prereq[i][1])
                   
                    if (matched[i][0]==roles[j]):
                        y[j]+=matched[i][1]
#             print(y)

            for i in range(len(y)):
                #check company and years
                #print(matched)
                for j in matched:
                    if (y[i]>=self.prereq[roles[i]][1]) and (j[2] not in self.black):
#                         print("Job assigned: ",roles[i])
                        self.assignJob(roles[i],i,env,name)
                        break
        
        
        self.printStatus(env,roles)                    
                        
                

    #more actions
    def assignJob(self,jname,pos,env,name):
        # decrement vacancy
        
        env.update_vacancy(pos)
        env.update_status(jname)
        #send email
        role=""
        if jname=='SA':
            role = "Software Analyst"
        if jname=='SSA':
            role = "Senior Software Analyst"
        if jname=='PL':
            role = "Program Lead"
        if jname=='AVP':
            role = "Assistant Vice President"
        if jname=='VP':
            role = "Vice President"
            
        if self.stop[pos]==0:
            print("No more vacancies for the job of ",role,"!!\n")
        else:
            print("Sending Email to candidate \' ",name," \' selected for __",role,"__ position!\n")

        
    def isFilled(self):
        s= 0
        for i in range(len(self.stop)):
            s+=self.stop[i]
        if(s==0):
            print("\nAll positions filled!! No more applications accepted..\n")
            return True
        else:
            return False
        
        
    def printStatus(self,env,roles):
        print("\n...Job roles vacancies at the moment...")
        for i in range(len(self.stop)):
            print(roles[i]," : ",self.stop[i])
```

```python
env = JobPortal()
agent = ModelAgent(env)

print("\n___Job portal agent___")
#take input
count = 1
while not agent.isFilled():
    userinput = []
    print("\nCandidate No. ",count)
    print("\nEnter your previous experience details...")
    n = int(input("How many previous positions did you have? "))
    name = input("How are you called? ")
    for i in range(n):
        li = []
        print("\nEnter details of job ",i+1)
        li.append(int(input("Enter number of working years: ")))
        li.append(input("Enter designation: "))
        li.append(input("Enter name of company: "))
        userinput.append(li)
    # print(userinput)
    agent.filter(env, name,userinput)
    count+=1
```
